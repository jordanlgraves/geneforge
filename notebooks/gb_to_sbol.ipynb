```python
import os
import sbol3

file_paths = [
    #'data/jbei/synberg/scraped/gb/DclpX_cassette.gb',
                'data/jbei/synberg/scraped/gb/jtk2919.gb',
                'data/jbei/synberg/scraped/gb/PY404.gb']

out_dir = 'data/jbei/synberg/sbol/'
os.makedirs(out_dir, exist_ok=True)
# convert to SBOL
for file_path in file_paths:
    print(file_path)
    fname = os.path.basename(file_path)
    outname = fname.replace('.gb', '.sbol')
    
    doc = sbol3.Document()
    doc.importFromFormat(file_path, True)
    doc.write(os.path.join(out_dir, outname))
```
